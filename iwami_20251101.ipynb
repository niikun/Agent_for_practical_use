```python
from collections import Counter, defaultdict

def build_transition_graph(
    df: pd.DataFrame,
    start_date: str,
    end_date: str,
    use: str = "name_code",
    collapse_consecutive: bool = True, # 同一地点の連続を圧縮
    min_weight: int = 1,               # この重み未満のエッジは除外
):
    """
    返り値:
      G: nx.DiGraph (ノード=地点, エッジ重み=遷移回数)
      edge_df: エッジ一覧 DataFrame [src, dst, weight]
    """
    df = df.copy()

    # 1) 日時をdatetime化
    if not pd.api.types.is_datetime64_any_dtype(df["occurredAtJst"]):
        df["occurredAtJst"] = pd.to_datetime(df["occurredAtJst"])

    # 2) 期間フィルタ（endは翌日の0時未満で含める）
    start = pd.to_datetime(start_date)
    end_next = pd.to_datetime(end_date) + pd.Timedelta(days=1)
    df = df[(df["occurredAtJst"] >= start) & (df["occurredAtJst"] < end_next)]

    if df.empty:
        raise ValueError("指定条件に一致するデータがありません。")

    # 3) 1地点=1アドレスの前提で、ノード属性に両方を保持（あとでラベル切替しやすく）
    #    期間内の name_code→contractAddress の対応を作る（重複があっても最初のを採用）
    name_to_addr = (
        df.dropna(subset=["name_code", "contractAddress"])
          .drop_duplicates(subset=["name_code"])    # nameは1対1の前提
          .set_index("name_code")["contractAddress"]
          .to_dict()
    )

    # 4) walletごとの時系列ソート → 連続遷移をカウント
    edge_counter = Counter()
    wallets = sorted(df["walletNo"].unique())
    for w in wallets:
        sub = df[df["walletNo"] == w].sort_values("occurredAtJst")
        seq = sub[use].tolist()
        if not seq:
            continue

        if collapse_consecutive:
            comp = [seq[0]]
            for s in seq[1:]:
                if s != comp[-1]:
                    comp.append(s)
            seq = comp

        # 連続ペア（u->v）
        edges = list(zip(seq[:-1], seq[1:]))
        edge_counter.update(edges)

    # 5) 最小重みでフィルタ
    if min_weight > 1:
        edge_counter = Counter({e: w for e, w in edge_counter.items() if w >= min_weight})

    # 6) グラフ描画
    G = nx.DiGraph()
    # ノード追加（属性で両方持つ）
    nodes = set()
    for (u, v) in edge_counter.keys():
        nodes.add(u)
        nodes.add(v)
    for n in nodes:
        G.add_node(n, name="name_code")

    # エッジ追加（重み）
    for (u, v), w in edge_counter.items():
        G.add_edge(u, v, weight=w)

    # 8) エッジ一覧
    edge_df = (
        pd.DataFrame([(u, v, d["weight"]) for u, v, d in G.edges(data=True)],
                     columns=["src", "dst", "weight"])
        .sort_values(["weight", "src", "dst"], ascending=[False, True, True])
        .reset_index(drop=True)
    )

    return G, edge_df
```

### Graph描画

```python
G_all, edges_all = build_transition_graph(
    df,
    start_date="2025-03-26", # 期間を指定（例：直近3か月）
    end_date="2025-06-30",
    use="name_code",          # or "contractAddress"
    collapse_consecutive=True,
    min_weight=50             # 弱いエッジを間引く（50以上で設定）
)

# Position を2次元にmapping
pos = {
    1:(20,65),
    2:(5,65),
    3:(0,0),
    4:(19,-8),
    5:(15,5),
    6:(20,15),
    7:(25,15),
    8:(20,35),
    9:(15,50),
    10:(40,65),
    11:(45,35)
}

# 円形にmappingする場合
# pos = nx.circular_layout(G_all)

plt.figure(figsize=(11, 9))
nx.draw_networkx_nodes(G_all, pos, node_size=600, node_color="lightgray")
nx.draw_networkx_labels(G_all, pos, font_size=16, font_color="black")

weights = [d["weight"] for _,_,d in G_all.edges(data=True)]
widths  = [1 + 5*(w-min(weights))/(max(weights)-min(weights)) if max(weights)>min(weights) else 3 for w in weights]

nx.draw_networkx_edges(G_all, pos, arrows=True, arrowstyle="-|>", arrowsize=14, width=widths)
nx.draw_networkx_edge_labels(G_all,
                             pos=pos,
                             edge_labels={(u,v): d["weight"] for u,v,d in G_all.edges(data=True)}, font_size=8)

plt.title("All users: directed weighted Graph")
plt.axis("off"); plt.tight_layout(); plt.show()
```
